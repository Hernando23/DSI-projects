## Step 1: Description

The dataset contains SAT scores for Math and Verbal tests for students of 45 of US states. The score values are average per state and the column 'Rate' contains the rate of participation.

##### Data dictionary for the dataset:

**Category** | **Description** | **Expected dtype**
--- | --- | --- | ---
State | US state acronym | String/object 
Rate | Participation rate | integer
Verbal | Average verbal test score | integer
Math | Average math test score |integer

##### Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def compute_std(df):
    return [np.std(df[column]) for column in df if df[column].dtype == 'int64']
```

```python
print compute_std(sat_scores)
```

## Step 4: Visualize the data

##### Using MatPlotLib and PyPlot, plot the distribution of the Rate using histograms.

```python
plt.hist(sat_scores['Rate'], bins = 10, color='r')
plt.title("Participation Rate Distribution")
plt.xlabel("Rate")
plt.ylabel("Frequency")
plt.show()
```

##### Plot the Math distribution

```python
plt.hist(sat_scores['Math'], bins = 10, color='b')
plt.title("Math Test Scores Distribution")
plt.xlabel("Math Test Score")
plt.ylabel("Frequency")
plt.show()
```

##### Plot the Verbal distribution

```python
plt.hist(sat_scores['Verbal'], bins = 10, color='y')
plt.title("Verbal Test Scores Distribution")
plt.xlabel("Verbal Test Score")
plt.ylabel("Frequency")
plt.show()
```

##### What is the typical assumption for data distribution?

The assumption of normality.

##### Does that distribution hold true for our data?

Judging by the charts, only histogram for math scores looks close to normal distrubution. The other two plots look like they are not normally distributed. 

We can conduct a test to check if the distribution is normal, i.e. if it's pvalue is smaller than 0.05:

```python
pvalue_rate = stats.normaltest(sat_scores['Rate'])[1]
pvalue_verbal = stats.normaltest(sat_scores['Verbal'])[1]
pvalue_maths = stats.normaltest(sat_scores['Math'])[1]
```

```python
def is_normal(pvalue):
    if pvalue > 0.05:
        print "This is a normal distribution"
    else:
        print "This is not normal distribution"
```

Test for participation rate:

```python
is_normal(pvalue_rate)
```

Test for verbal scores:

```python
is_normal(pvalue_verbal)
```

Test for math scores:

```python
is_normal(pvalue_maths)
```

Which confirms our suspicions.

##### Plot some scatterplots. **BONUS**: Use a PyPlot `figure` to present multiple plots at once.

```python
plt.figure(figsize=(8,6))

plt.subplot(221)
plt.scatter(sat_scores['Rate'], sat_scores['Verbal'], c='r')
plt.title('Rate vs Verbal')
plt.grid(True)
plt.xlabel("Rate")
plt.ylabel("Verbal")

plt.subplot(222)
plt.scatter(sat_scores['Rate'], sat_scores['Math'], c='g')
plt.title('Rate vs Maths')
plt.grid(True)
plt.xlabel("Rate")
plt.ylabel("Maths")


plt.subplot(223)
plt.scatter(sat_scores['Verbal'], sat_scores['Math'])
plt.title('Verbal vs Maths')
plt.grid(True)
plt.xlabel("Verbal")
plt.ylabel("Maths")

plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.35,
                    wspace=0.25)

plt.show()
```

##### Are there any interesting relationships to note?

There seems to be a very clear positive correlation between Verbal and Math scores, meaning that state with high Verbal score is likely to have high Math score. 

It also looks like there is a negative correlation between Rate of Participation and Verbal scores and also a bit less pronounced negative correlation between Rate of Participation and Math scores. Both quite surprisingly indicate that states with higher rate of participation are likely to have worse test scores.
